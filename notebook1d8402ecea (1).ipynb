```python
# Feature 1. High = candidate for Карта для путешествий 

# from transactions: (Авто + АЗС + Такси + Отели + Путешествия)/total spend



# // need to check if user changes cities frequently????

# Feature 2. High = Премиальная карта

# from transactions: (Кафе и рестораны + Ювелирные украшения + Косметика и Парфюмерия)/total spend

# status: premium client

# Feature 3. High = Кредитная карта 

# from transactions: (Кино + Едим дома + Смотрим дома + Играем дома)/total_spend

# from transfers: cc_repayment_out + installment_payment_out
# status: student (Кешбэк: до 10% в трёх «любимых категориях», выбираете каждый месяц  и на 10% онлайн услуги: игры, доставка, кино)

# Feature 4. High = Обмен валют

# from transfers: fx_buy + fx_sell

# Feature 5. High = Кредит наличными

# from transfers: total_out/total_in if > 1 then high probability to offer Кредит наличными

# Feature 6. High = Депозит Сберегательный / Накопительный

# avg_balance – monthly_spend

# Feature 7.  High = Депозит Мультивалютный

# merge feature 6 and feature 4 if both are high then feature 7 is most likely to also be suitable for client

# Feature 8.  High = Инвестиции

# from transfers: invest_in + invest_out
# Feature 9.  High = Золотые слитки

# from transactions: gold_sell_in + gold_buy_out
```

```python
import glob, os
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score


# Paths
transaction_files = glob.glob("/kaggle/input/decentrathon/case 1/*_transactions_3m.csv")
transfer_files = glob.glob("/kaggle/input/decentrathon/case 1/*_transfers_3m.csv")

client_features = []

for tfile in transaction_files:
    client_code = os.path.basename(tfile).split("_")[1]   # e.g., "10"
    tfiles = f"/kaggle/input/decentrathon/case 1/client_{client_code}_transactions_3m.csv"
    trfile = f"/kaggle/input/decentrathon/case 1/client_{client_code}_transfers_3m.csv"
    
    # Read
    tx = pd.read_csv(tfiles)
    tr = pd.read_csv(trfile)

    # Total spend
    total_spend = tx['amount'].sum()

    # --- Feature 1: Travel card ---
    travel_cats = ['Авто','АЗС','Такси','Отели','Путешествия']
    travel_spend = tx[tx['category'].isin(travel_cats)]['amount'].sum()
    feat_travel = travel_spend / total_spend if total_spend > 0 else 0

    # --- Feature 2: Premium card ---
    premium_cats = ['Кафе и рестораны','Ювелирные украшения','Косметика и Парфюмерия']
    prem_spend = tx[tx['category'].isin(premium_cats)]['amount'].sum()
    feat_premium = prem_spend / total_spend if total_spend > 0 else 0
    is_premium = int("Премиальный клиент" in tx['status'].unique())

    # --- Feature 3: Credit card ---
    cc_cats = ['Кино','Едим дома','Смотрим дома','Играем дома']
    cc_spend = tx[tx['category'].isin(cc_cats)]['amount'].sum()
    cc_transfers = tr[tr['type'].isin(['cc_repayment_out','installment_payment_out'])]['amount'].sum()
    is_student = int("студент" in tx['status'].unique())
    feat_cc = (cc_spend / total_spend if total_spend > 0 else 0) + cc_transfers

    # --- Feature 4: FX ---
    feat_fx = tr[tr['type'].isin(['fx_buy','fx_sell'])]['amount'].sum()

    # --- Feature 5: Cash credit ---
    total_in = tr[tr['direction']=="in"]['amount'].sum()
    total_out = tr[tr['direction']=="out"]['amount'].sum()
    feat_cash_credit = total_out / total_in if total_in > 0 else 0

    # --- Feature 6: Savings ---
    feat_saving = total_in - total_spend

    # --- Feature 7: Multi-currency deposit ---
    feat_multicurrency = feat_saving * feat_fx

    # --- Feature 8: Investments ---
    feat_invest = tr[tr['type'].isin(['invest_in','invest_out'])]['amount'].sum()

    # --- Feature 9: Gold ---
    feat_gold = tr[tr['type'].isin(['gold_sell_in','gold_buy_out'])]['amount'].sum()

    client_features.append({
        "client_code": client_code,
        "travel": feat_travel,
        "premium": feat_premium + is_premium,
        "credit_card": feat_cc + is_student,
        "fx": feat_fx,
        "cash_credit": feat_cash_credit,
        "saving": feat_saving,
        "multicurrency": feat_multicurrency,
        "invest": feat_invest,
        "gold": feat_gold
    })

# --- Build DataFrame ---
features = pd.DataFrame(client_features).set_index("client_code")

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

# KMeans
kmeans = KMeans(n_clusters=10, random_state=42)
features['cluster'] = kmeans.fit_predict(X_scaled)

# --- Visualize cluster centers ---
centers = pd.DataFrame(kmeans.cluster_centers_, columns=features.columns[:-1])
plt.figure(figsize=(12,6))
sns.heatmap(centers, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Cluster centers (scaled feature values)")
plt.show()

labels = features['cluster']

sil_score = silhouette_score(X_scaled, labels)
db_score  = davies_bouldin_score(X_scaled, labels)
ch_score  = calinski_harabasz_score(X_scaled, labels)

print(f"Silhouette Score: {sil_score:.3f}")
print(f"Davies-Bouldin Index: {db_score:.3f}")
print(f"Calinski-Harabasz Index: {ch_score:.1f}")
```

```python
import pandas as pd, numpy as np, re
from datetime import datetime
import calendar

RU_MONTHS = {1:"январе",2:"феврале",3:"марте",4:"апреле",5:"мае",6:"июне",7:"июле",8:"августе",9:"сентябре",10:"октябре",11:"ноябре",12:"декабре"}

TRAVEL_CATS  = ['Авто','АЗС','Такси','Отели','Путешествия']
PREMIUM_CATS = ['Кафе и рестораны','Ювелирные украшения','Косметика и Парфюмерия']
CC_CATS      = ['Кино','Едим дома','Смотрим дома','Играем дома']

def _last_month_phrase(tx, tr):
    dates = []
    if 'date' in tx: dates += pd.to_datetime(tx['date'], errors='coerce').dropna().tolist()
    if 'date' in tr: dates += pd.to_datetime(tr['date'], errors='coerce').dropna().tolist()
    if not dates: return None
    m = max(dates).month
    return f"в {RU_MONTHS[m]}"

def build_observed_behavior(tx: pd.DataFrame, tr: pd.DataFrame, city: str | None, candidate_products=None):
    """
    Возвращает релевантное наблюдение, завязанное только на продукты кластера.
    candidate_products: список словарей продуктов (title, benefits, fit_for)
    """
    when = _last_month_phrase(tx, tr)

    # категории
    share = lambda cats: (tx.loc[tx['category'].isin(cats), 'amount'].sum() /
                          (tx['amount'].sum() or 1.0)) if not tx.empty else 0.0
    s_travel  = share(TRAVEL_CATS)
    s_premium = share(PREMIUM_CATS)
    s_cc      = share(CC_CATS)

    fx_amt = tr.loc[tr.get('type').isin(['fx_buy','fx_sell']), 'amount'].sum() if not tr.empty else 0
    in_amt = tr.loc[tr.get('direction')=="in",  'amount'].sum() if not tr.empty else 0
    out_amt= tr.loc[tr.get('direction')=="out", 'amount'].sum() if not tr.empty else 0

    signals = []
    if candidate_products:
        titles = [p["title"] for p in candidate_products]

        if any("Путешествий" in t for t in titles) and s_travel >= 0.25:
            signals.append("частые поездки и такси")
        if any("Премиальная" in t for t in titles) and s_premium >= 0.25:
            signals.append("много ресторанов и премиальных покупок")
        if any("Кредитная карта" in t or "онлайн" in " ".join(p["fit_for"]) for p in candidate_products) and s_cc >= 0.25:
            signals.append("онлайн-развлечения и доставки")
        if any("Обмен валют" in t or "Золотые" in t or "Депозит мультивалютный" in t for t in titles) and fx_amt > 0:
            signals.append("обмен валюты в приложении")
        if any("Кредит" in t for t in titles) and out_amt/in_amt>1.2 if in_amt>0 else False:
            signals.append("переводы чаще, чем поступления")

    city_part = f" в {city.lower()}" if city else ""
    if when and signals:
        return f"{when}{city_part} — {signals[0]}"
    if when:
        return f"{when}{city_part}"
    return signals[0] if signals else "наблюдаем активность по счету"
```

```python
import os, json, time, re
import pandas as pd
from pathlib import Path

# ---------- tiny helpers ----------
def safe_mode(series):
    try:
        s = series.dropna()
        return None if s.empty else s.mode().iat[0]
    except Exception:
        return None

def load_client_frames(cid: str):
    base = "/kaggle/input/decentrathon/case 1"
    tx_path = f"{base}/client_{cid}_transactions_3m.csv"
    tr_path = f"{base}/client_{cid}_transfers_3m.csv"
    tx = pd.read_csv(tx_path) if os.path.exists(tx_path) else pd.DataFrame(columns=["date","category","amount","name","city","status"])
    tr = pd.read_csv(tr_path) if os.path.exists(tr_path) else pd.DataFrame(columns=["date","type","direction","amount","name","city","status"])
    return tx, tr

def client_profile_from_frames(tx: pd.DataFrame, tr: pd.DataFrame):
    name  = safe_mode(tx["name"])  if "name"  in tx.columns and not tx.empty else (safe_mode(tr["name"])  if "name"  in tr.columns and not tr.empty else None)
    city  = safe_mode(tx["city"])  if "city"  in tx.columns and not tx.empty else (safe_mode(tr["city"])  if "city"  in tr.columns and not tr.empty else None)
    status= safe_mode(tx["status"])if "status"in tx.columns and not tx.empty else (safe_mode(tr["status"])if "status"in tr.columns and not tr.empty else None)
    return name, city, status

# Rate-limit aware single call (uses server-provided retry_delay if present)
def generate_with_backoff(parts, max_retries=4):
    delay = 1.0
    last_err = None
    for _ in range(max_retries):
        try:
            resp = gmodel.generate_content(parts)
            txt = (resp.text or "").strip()
            if txt:
                return txt
        except Exception as e:
            last_err = e
            msg = str(e)
            m = re.search(r"retry_delay\s*\{\s*seconds:\s*(\d+)", msg)
            if "429" in msg:
                wait = int(m.group(1)) + 1 if m else int(delay)
                time.sleep(wait)
                delay = min(delay * 1.7, 20)
                continue
            time.sleep(delay)
            delay = min(delay * 1.7, 10)
    if last_err:
        raise last_err
    return ""

# ---------- main loop ----------
rows = []
clusters = sorted(features["cluster"].unique().tolist())
print("Clusters:", clusters)

for cl_id in clusters:
    sub = features[features["cluster"] == cl_id]
    product_title = CLUSTER_TO_PRODUCT_TITLE[int(cl_id)]

    for cid, _ in sub.iterrows():
        # Load client data & profile
        tx, tr = load_client_frames(str(cid))
        name, city, status = client_profile_from_frames(tx, tr)

        # Build Option-A prompt → one-line push
        parts = build_prompt_v2(str(cid), tx, tr, int(cl_id), name, city, status)
        try:
            push_text = generate_with_backoff(parts)
        except Exception as e:
            # deterministic fallback if model call fails hard
            push_text = f"{name+', ' if name else ''}заметили ваши операции. {product_title}: отличные условия для вас. Открыть"

        rows.append({
            "client_code": str(cid),
            "client_name": name or "",
            "product_title": product_title,
            "push": push_text
        })

print(f"Generated pushes: {len(rows)}")

# ---------- save ----------
out_path = "/kaggle/working/push_texts.json"
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(rows, f, ensure_ascii=False, indent=2)

print("Saved:", out_path)
```
